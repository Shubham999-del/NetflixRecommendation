```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings=pd.read_csv("/content/drive/MyDrive/Netflix dataset/ratings.csv")
ratings.head()
```

```python
ratings.shape
```

```python
movies = pd.read_csv("/content/drive/MyDrive/Netflix dataset/movies.csv")
movies.head()
```

```python
data = pd.merge(ratings,movies,right_on='movieId',left_on='movieId',how='inner')
```

```python
data.head()
```

```python
# total unique genres
genres = []
movie_genres = []
for genre in data['genres']:
  if '|' in genre:
    movie_genres = genre.split('|')
    for movie_genre in movie_genres:
      if movie_genre not in genres:
        genres.append(movie_genre)
```

```python
len(genres)
```

```python
genres
```

```python
genres_summary = []
```

```python
for genre in genres:
  temp_df = data[data['genres'].str.contains(genre)]
  genre_summary = {}
  genre_summary['genre'] = genre
  genre_summary['total_ratings'] = temp_df.shape[0]
  genre_summary['Mean Ratings'] = np.round(temp_df['rating'].mean(), decimals = 2)
  genres_summary.append(genre_summary)
```

```python
genres_summary
```

```python
data_genre_summary = pd.DataFrame(genres_summary)
```

```python
data_genre_summary
```

```python
##top 5 Most popular genre
data_genre_summary.sort_values('total_ratings' ,ascending=False)[['genre','total_ratings']][:5]
```

```python
## Top rated genres
data_genre_summary.sort_values('Mean Ratings',ascending = False)[['genre','Mean Ratings']][:5]
```

```python
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate
```

```python
reader = Reader()
model = SVD()
```

## lets do this for user id = 1

```python
data.columns
```

```python
data
```

```python
movie_recommendations = []
```

```python
for genre in genres:
  reader = Reader()
  model = SVD()
  temp_df = data[data['genres'].str.contains(genre)]
  temp_data = Dataset.load_from_df(temp_df[['userId','movieId','rating']],reader)
  cross_validate(model, temp_data, measures=['RMSE'], cv=3)

  # recommendation for user 1
  user_1 = temp_df[['movieId', 'title', 'genres']].copy()
  user_1.reset_index()

  # estimated scores
  user_1['Estimated_Score'] = user_1['movieId'].apply(lambda x : model.predict(1,x).est)
  user_1 = user_1.sort_values('Estimated_Score', ascending=False)

  user_1 = user_1.drop_duplicates()

  # top recommended movie for this genre
  highest_recommended_movie = user_1.iloc[0]['title']

  recommendation = {}
  recommendation['genre'] = genre
  recommendation['Movie Name'] = highest_recommended_movie

  movie_recommendations.append(recommendation)
```

```python
movie_recommendations
```

```python
recommendations_for_user_1 = pd.DataFrame(movie_recommendations)
```
